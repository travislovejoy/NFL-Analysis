##                       Analyzing the 2012 San Francisco 49ers QB Change

The 2012 San Francisco 49ers began the season with Super Bowl aspirations, after a heart breaking loss in the 2011 NFC Championship games. The 49ers opened the season with Alex Smith commanding the offense to an impressive 6-2 record to begin the season. In week game 10 against the St. Louis Rams, Alex Smith suffered a concussion forcing him to be pulled from the game, as well as the 49ers week 11 mathup against the Chicago Bears. This allowed backup Quarterback Colin Kaepernick an opportunity to run the 49ers offense. Kaepernick took full advantage of this opportunity and never relinquished the starting role back to Alex Smith, and ultimatley led the 49ers to a Super Bowl appearnce where the 49ers came up just short of the Lombardi Trophy.

To me this is one of the more interesting football stories in the last decade for a variety of reasons, and not solely for the reason that I am a 49er fan. One of the reasons this is such an interesting story is that no backup quaterback has lead his team to a Super Bowl since Tom Brady in 2001. Additionally there has long been a unwritten rule that a player cannot lose his job due to injury. Becuase these reasons, the 49ers QB change is an interesting situation needing further examination, which I will do in this notebook.

For the purposes of this notebook I will only consider Regular seasons games in which either Quaterback played a full game. Meaning that Playoff matchups as well as the Week 10 matchup in which Kaepernick replaced Smith mid game will be disregarded for this analysis. Alex Smith stats will include weeks 1-8 while Colin Kaepernick stats will include weeks 11-17. 

## Player Stat Comparison

```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as stats
```

```python
Smith = pd.read_csv('/home/travis/NFL/Analysis/QBchange/Data/AlexSmith.csv')

Smith
```

```python
Kaepernick=pd.read_csv('/home/travis/NFL/Analysis/QBchange/Data/ColinKaepernick.csv')

Kaepernick
```

```python
combo= Smith.append(Kaepernick)
combo['WinPercentage']= combo['win']*100
combo=combo.drop('win')
combo=combo.groupby(['full_name'])
combo=combo[['comp','pass_att', 'pass_yards', 'pass_tds', 'int', 'sacked', 'rush_att', 'rush_yards', 'rush_tds', 'WinPercentage']].mean()
combo=combo.round(1)
combo.transpose()
```

The game averages of the two players are very similar. With the only real differences being the increase of passing and rushing yards by Colin Kaepernick. The rushing yards are expected given Colin Kaepernick being labeled a running quarterback but the increase in passing yards is interesting.

## Playcalling

Next we will look at wether the quaterback change lead the 49ers coaching staff to make adjustments in thier playcall (run/pass) ratio. 

```python
PlayCalling= pd.read_csv('/home/travis/NFL/Analysis/QBchange/Data/PlayCalling.csv')
```

```python
Smith= PlayCalling[PlayCalling['week']<8]
Kaepernick= PlayCalling[PlayCalling['week']>8]
#niners=PlayCalling[PlayCalling['team_id']=='SF']
```

```python
#Remove opp;onent playcalling and display Sanfrancisco offensive playcalling while Alex Smith at QB
SmithOff=Smith[Smith['team_id']=='SF']
SmithOff
```

```python
#Do Same for Kaepernick
KaepernickOff=Kaepernick[Kaepernick['team_id']=='SF']
KaepernickOff
```

```python
#Calculate Averages and S.D. with Alex Smith at QB
Smithpass = np.mean(SmithOff['pass'])
Smithrush= np.mean(SmithOff['rush'])
SmithpassSTD = np.std(SmithOff['pass'])
SmithrushSTD= np.std(SmithOff['rush'])
SmithPos= np.mean(SmithOff['pos_time'])
SmithPosSTD= np.std(SmithOff['pos_time'])


print Smithpass, SmithpassSTD, Smithrush, SmithrushSTD, SmithPos, SmithPosSTD
```

```python
Kaepernickpass = np.mean(KaepernickOff['pass'])
Kaepernickrush= np.mean(KaepernickOff['rush'])
KaepernickpassSTD = np.std(KaepernickOff['pass'])
KaepernickrushSTD= np.std(KaepernickOff['rush'])
KaepernickPos= np.mean(KaepernickOff['pos_time'])
KaepernickPosSTD= np.std(KaepernickOff['pos_time'])

print Kaepernickpass, KaepernickpassSTD,Kaepernickrush, KaepernickrushSTD, KaepernickPos, KaepernickPosSTD
```

```python
%matplotlib nbagg
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
passPlays= [Smithpass, Kaepernickpass]
passSTD=[SmithpassSTD, KaepernickpassSTD]
rushPlays= [Smithrush, Kaepernickrush]
rushSTD= [SmithrushSTD, KaepernickrushSTD]
ind= np.arange(len(passPlays))
barwidth=.4

rRect=ax.bar( ind,rushPlays, color='#E6BE8A', label= "Rush Plays", width=barwidth, yerr= rushSTD, error_kw=dict(ecolor='black', lw=.5, capsize=5,capthick=2))
pRect=ax.bar( ind+barwidth,passPlays, color='#AF1E2C',label= "Pass Plays", width=barwidth, yerr= passSTD, error_kw=dict(ecolor='black', lw=.5, capsize=5,capthick=2))
#ax.bar( np.arange(2),rushPlays, bottom= passPlays, color='gold', label= "Pass Plays", width=barwidth)

ax.set_xticks(ind+ barwidth)
ax.set_xticklabels(('Alex Smith', 'Colin Kaepernick'))

def labeler(rects):
        for rect in rects:
                
                height= rect.get_height()
                ax.text(rect.get_x()+rect.get_width()/2. , height, '%.1f' % float(height), ha='center', va='bottom', fontweight='bold', fontsize=15)
                
labeler(pRect)
labeler(rRect)

ax.legend(loc=1)
ax.set_ylabel('# of Plays')
ax.set_title('Run/Pass Ratio Comparison')
```

As we can see in the above graph, the playcalling of the 49ers remained rougly the same, as the 49ers were able to roughly obtain an 50/50 split for run pass playcalling. Kaepernick did increase the run play total slightly, however
from database given we are not able to decipher which of these are called runs rather than QB scrambles.

## Offensive Yardage Breakdown

```python
fig, ax = plt.subplots()
ind= np.arange(len(passingyards))
barwidth=.2

rRect=ax.bar( ind,rushingyards, color='#E6BE8A', label= "Rush Yards", width=barwidth, yerr= rushingyardsSTD, error_kw=dict(ecolor='grey', lw=.5, capsize=5,capthick=2))
pRect=ax.bar( ind+barwidth,passingyards, color='#AF1E2C',label= "Pass Yards", width=barwidth, yerr= passingyardsSTD, error_kw=dict(ecolor='grey', lw=.5, capsize=5,capthick=2))
tRect=ax.bar( ind+barwidth*2 ,totalyards, color='white',label= "Total Yards", width=barwidth, yerr= totalyardsSTD, error_kw=dict(ecolor='grey', lw=.5, capsize=5,capthick=2))

#ax.bar( np.arange(2),rushPlays, bottom= passPlays, color='gold', label= "Pass Plays", width=barwidth)f

ax.set_xticks(ind+ barwidth*3 /2)
ax.set_xticklabels(('Alex Smith', 'Colin Kaepernick'))

def labeler(rects):
        for rect in rects:
                
                height= rect.get_height()
                ax.text(rect.get_x()+rect.get_width()/2. , height, '%.1f' % float(height), ha='center', va='bottom', fontweight='bold', fontsize=15)
                
labeler(pRect)
labeler(rRect)
labeler(tRect)

ax.legend(loc=0)
ax.set_ylabel('Yards')
ax.set_title('Passing and Rushing Yards Comparison')
```

As we can see here a Smith led offense resulted in higher team rushing yards, and overall yards, while Kaepernick led offense had slight increase of pass yards over an Alex Smith ran offense. 

Another thing to note is that the error bars are greater in a Smith run offense. This may be explained by an outlier in the the Smith data set. Looknig at the above table shows that Week 5 in which the 49ers gained over 600 yards, against an overmatched Buffalo team. This appears be an outlier, causing inflated yardage totals for Smith. In a later section I will remove this game from the dataset to achieve a more fair comparison of the players the players.

## Conclusion

AS the above analysis shows a 49er run offense by either of the Quaterbacks resulted in similar results. From our analysis it seems that Alex Smith's numbers may have been inflated by a Week 5 matchup against the Buffalo Bills in which he threw for 6 touchdowns and helped the 49er offense amass over 600 yards. Removing this outlier from the analysis allowed Colin Kaepernick to perform much more favorbly, and helps to justify 49ers coach Jim Harbaugh decision to allow Colin Kaepernick to remain as the starting Quarterback.

In the short term Jim Harbaugh decision was vindicated as Colin Kaepernick led the 49ers to the Super Bowl, in which the came up yards short of the Lombardi Trophy, and the numbers do not show anything to prove that an Alex Smith run offesne would have yielded better results.

After the season Alex Smith was traded to the Kansas City Chiefs, and Colin Kaepernick was heralded as the quarterback of the future. Unfortunatley for the 49ers, Colin Kaepernick was never able to lead the 49ers back to the Super Bowl, allowing the debate of wether the 49ers made the correct decision to continue.
